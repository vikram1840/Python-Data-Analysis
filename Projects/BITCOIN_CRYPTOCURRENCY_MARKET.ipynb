```python
# Importing pandas
import pandas as pd

# Importing matplotlib and setting for plotting 
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')

# Reading datasets/coinmarketcap_06122017.csv into pandas
data = pd.read_csv(r"C:\Users\vikramku\Desktop\Target_Data_Analyst\Python\PROJECTS\Project _Bitcoin\Exploring the Bitcoin Cryptocurrency Market\datasets\coinmarketcap_06122017.csv")

# Printing out the first few lines
data.head()
```

```python
# Selecting the columns 'id' and the 'market_cap_usd' and assign to dataframe market_cap_raw
market_cap_raw =data[['id','market_cap_usd']]
market_cap_raw
```

```python
# Counting the number of values
market_cap_raw.count()
```

```python
# As the above count suggest that some of the cryptocurrencies not having market capitalization so better to discard them

## First method to discard them is drop all rows having market_cap_usd value NaN
market_cap_raw.dropna()

## Second method by applying condition where market_cap_usd>0
cap=market_cap_raw[market_cap_raw['market_cap_usd']>0]

## Another method to apply condition like market_cap_usd>0 bu using query() method
cap = market_cap_raw.query('market_cap_usd>0')
cap
```

```python
# Now count the values again and confirm we have only cryptocurrencies having some market capitalization
cap.count()
```

```python
# To Select and visualize market capitalization of the top 10 cryptocurrencies

## First sort values based on market_cap_usd in case it is not ordered in descending order
cap.sort_values(by=['market_cap_usd'],ascending=False)

# Setting the column 'id' as index
cap_new=cap.set_index('id')

# Calculate the percentage of market capitalization for each coin and 'assign()' them to cap_new
cap_new=cap_new.assign(market_cap_perc=round((cap_new['market_cap_usd']/cap_new['market_cap_usd'].sum())*100,2))

## Another method to use 'assign()' with lambda function and calculate market_cap_perc
cap_new = cap_new.assign(market_cap_perc=lambda x:(x.market_cap_usd/cap_new.market_cap_usd.sum())*100)

#Selecting Top 10 cryptocurrencies as per their %market_capitalization
cap10=cap_new.head(10)


```

```python
# visualize market capitalization of the top 10 cryptocurrencies

## Declare Titae and Ylabel for plot
TOP_CAP_TITLE= 'Top 10 market capitalization'
TOP_CAP_YLABEL = '% of total cap'

# Plotting the barplot with the title defined above
ax= cap10.plot.bar(y='market_cap_perc',title='Top 10 market capitalization')

# Annotating the y axis with the label defined above
ax.set_ylabel('% of total cap')

# Annotating X-lable does not make any sense so will remove x-lable
ax.set_xlabel('')
```

```python
# Create a dataframe thet contains the volatility information on cryptocurrencies

## Select column 'id', and 'percent_change in 7d and 24h column' and analysze the data
volatility= data[['id','percent_change_24h','percent_change_7d']]

## Select 'id' as index and dropped NaN rows
volatility= volatility.set_index('id').dropna()

## Sorting the DataFrame by percent_change_24h in ascending order
volatility = volatility.sort_values(by='percent_change_24h')

## Checking the first few rows
volatility.head()
```

```python
# plot the Top 10 biggest gainers and biggest losers 

## In First method we can plot Top gainers and losers in terms of 24h one by one and same we can do in terms of weekly(7d change)
## In second method we can perform all these just by calling function which will reducec coding

### Lets go with First method and plot Top 10 losers in left and gainers in right side

# Assign top 10 list to loers
#volatility_24h_losers=volatility['percent_change_24h'].head(10)
volatility_24h_losers=volatility['percent_change_24h'][:10]

# Assign last 10 as gainers
#volatility_24h_gainers=volatility['percent_change_24h'].tail(10)
volatility_24h_gainers=volatility['percent_change_24h'][-10:]

# Making the subplot and the figure for two side by side plots
fig,(ax1,ax2)  = plt.subplots(nrows=1,ncols=2,figsize=(10,6))

 # Plotting with pandas the barchart for the top 10 losers
ax1 = volatility_24h_losers.plot.bar(title="top losers",color='darkred',ax=ax1)

# Same as above, but for the top 10 winners
ax2 = volatility_24h_gainers.plot.bar(title="top winners",color='darkblue',ax=ax2)

# Setting the figure's main title to the text passed as parameter
fig.suptitle("24 hours top losers and winners")

# Setting the ylabel to '% change'
ax1.set_ylabel("% change")
```

```python
# In second Method to plot top 10 loosers and gainsers by calling function

def top10_subplot(volatility_top10,title):
    # Making the subplot and the figure for two side by side plots
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 6))
    
    # Plotting with pandas the barchart for the top 10 losers
    ax = volatility_top10[:10].plot.bar(title="top losers",color='darkred',ax=axes[0])
    
    # Setting the figure's main title to the text passed as parameter
    fig.suptitle(title)
    
    # Setting the ylabel to '% change'
    ax.set_ylabel("% change")
    
    # Same as above, but for the top 10 winners
    ax = volatility_top10[-10:].plot.bar(title="top winners",color='darkblue',ax=axes[1])
    
    # Returning this for good practice, might use later
    return fig, ax

TITLE = "24 hours top losers and winners"

# Calling the function above with the 24 hours period series and title DTITLE  
fig, ax = top10_subplot(volatility.percent_change_24h,TITLE)
```

```python
# Now check volatility as per weekly data basis

# First sort values of column 'percent_change_7d'
volatility7d= volatility.sort_values(by=['percent_change_7d'])

# Now call function top10_subplot with volatility7d
WTITLE="Weekly top losers and winners"
fig,ax =top10_subplot(volatility7d['percent_change_7d'],WTITLE)




```
