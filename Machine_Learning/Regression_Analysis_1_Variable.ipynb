```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn import linear_model
```

##### Most importnat libray is linear_model through which we will use linear regression

Load Data in pandas dataframe

```python
df= pd.read_excel(r"C:\Users\vikramku\Desktop\Target_Data_Analyst\Machine Learning\Liner_regression\Prices.xlsx")
df
```

##### Scatter plot to check whether a liner relation between area and price is there or not

```python
plt.scatter(df['Area'],df['price'],color='red',marker='+')
plt.xlabel('Area(Sqft)',size=20)
plt.ylabel('Price(US$)',size=20)
```

##### Another methos used seaborn library to check the same linear relationship

```python
plt.figure(figsize=(12,10))
sns.set()
sns.lmplot(x='Area',y='price',data=df)
```

```python
sns.scatterplot(x='Area',y='price',data=df)
```

##### implement linear regression model by creating an object for linearRegression and try best fit

```python
reg = linear_model.LinearRegression()
reg.fit(df[['Area']],df.price)
```

##### Predict the result from derived model

```python
reg.predict([[3300]])
```

#### Derive the value of slope and intercept used in linear regression formula- Y= MX+b

```python
reg.coef_
```

```python
reg.intercept_
```

#### Check The Score of Our LineraRegression Model

```python
reg.score(df[['Area']],df.price)
```

##### Now We Will perform one Excercise with Canada percapita income vs Year and will predict per capita income from year 2017-2020

##### To Import Canada Income excel and predict per capita income in year 2020

```python
income_data= pd.read_excel(r"C:\Users\vikramku\Desktop\Target_Data_Analyst\Machine Learning\Liner_regression\Canada_income.xlsx")
income_data
```

##### import linear_model library contains LinearRegression 

```python
from sklearn import linear_model
model= linear_model.LinearRegression()
```

```python
sns.lmplot(x='year',y='per capita income (US$)',data=income_data)
```

##### Best fit into regression model

```python
model.fit(income_data[['year']],income_data['per capita income (US$)'])
```

```python
model.predict([[2020]])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.score(income_data[['year']],income_data['per capita income (US$)'])
```

##### make the prediction from year 2017 to 2020

##### One method is simply pass the year in model and predict value and anther is pass those year in dataframe

```python
year1=[2017,2018,2019,2020]
#year2=np.array(year1)
```

```python
model.predict([[2017],[2018],[2019],[2020]])
```

Pass the year in dataframe

```python
year_df= pd.DataFrame({'year':year1})
year_df
```

```python
year_df['per capita income (US$)']=model.predict(year_df[['year']])
year_df
```

#### Concat this new dataframe with predicted value and concat them with hostorical data

```python
income_data= pd.concat([income_data,year_df],ignore_index=True)
income_data
```

```python
sns.set()
plt.figure(figsize=(15,10))
sns.lmplot(x='year',y='per capita income (US$)',data=income_data);
```
