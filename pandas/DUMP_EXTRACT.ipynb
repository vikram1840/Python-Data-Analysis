#### Code Written to import CM dump in pandas DataFrame and then Delimit column,Insert new column and assign them value

```python
import pandas as pd
df= pd.read_excel(r"C:\Users\vikramku\Desktop\Target_Data_Analyst\Python\Data Analysis\pandas\Utrancell.xlsx")
df.head()
```

##### Insert New Column 'RNCName' and 'UtranCell' at start of DataFrame

```python
for i in range(2):
     if i==0:
        column_name="RNCName"
     elif i==1:
        column_name="UtranCell"
     df.insert(i,column_name,'any')
df.head()
```

##### To extract RNCName from coulmn "MO" and assign to column "RNCName". Use series.str.extract function to extract RncName.

```python
df['RNCName']=df['MO'].str.extract("(RNC\w+)",expand=True)
df.head()
```

##### To extract UtranCell from column "MO" and assign to column "UtranCell". Use series.str.extract function to extract UtranCell Name

```python
df['UtranCell']=df['MO'].str.extract("(UtranCell=\w+)",expand=True)
df.head()
```

##### Replace text "UtranCell="  from column "Utrancell "

```python
df['UtranCell']=df['UtranCell'].str.replace('UtranCell=',"")
df.head()
```

#### Drop unwanted Column "MO" and "filename"

```python
df.drop(['MO','FileName'],axis=1,inplace=True)
df.head()
```
