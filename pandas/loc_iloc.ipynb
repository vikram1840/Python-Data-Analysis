```python
import pandas as pd
```

```python
ufo=pd.read_csv("http://bit.ly/uforeports")
ufo.head()
```

#### Start With loc

##### Row Selection

```python
ufo.loc[0,:] # zero rows and all column
```

```python
ufo.loc[[0,1,2],:] # to select different rows
```

```python
ufo.loc[0:2,:] # loc inclusive range that is 0 and 2 both 
```

##### Column Selection using loc

```python
ufo.loc[:,'City'].head() # single column
```

```python
ufo.loc[:,['City','Time']].head() # many column
```

```python
ufo.loc[:,'City':'Time'].head() # also can use range for column selection
```

```python
ufo.loc[0:2,'City':'State'] # select row and column both
```

```python
ufo.head(3).drop('Time',axis=1)
#ufo.head(3).drop(columns='Time')
```

```python
ufo[ufo.City=='Oakland'] 
```

```python
ufo.loc[ufo.City=='Oakland',:] # Same can be done using loc as well
```

```python
ufo['State'][ufo.City=='Oakland']
```

```python
ufo.loc[ufo.City=='Oakland','State'] # Same can be done using loc as well
```

##### Unlock iloc

```python
ufo.iloc[:,[0,3]].head()# access column with integer wise rather than name
```

```python
ufo.iloc[:,:].head()
```

```python
ufo.iloc[:,0:3].head() # 3 exclusive in column range
```

```python
ufo.iloc[0:3,:] # exclusive 3
```
